## PREPARAÇÃO DA ESTRUTURA

Responsável pela preparação da estrutura inicial dos CATALOGs, DATABASES/SCHEMAS.

```sql
%sql
DROP CATALOG IF EXISTS mvp_ed_ceap CASCADE;
CREATE CATALOG mvp_ed_ceap;
```

```sql
%sql
USE CATALOG mvp_ed_ceap;
```

```sql
%sql
DROP SCHEMA IF EXISTS staging CASCADE;
CREATE SCHEMA staging;

DROP SCHEMA IF EXISTS layer_bronze CASCADE;
CREATE SCHEMA layer_bronze;

DROP SCHEMA IF EXISTS layer_silver CASCADE;
CREATE SCHEMA layer_silver;

DROP SCHEMA IF EXISTS layer_gold CASCADE;
CREATE SCHEMA layer_gold;
```
