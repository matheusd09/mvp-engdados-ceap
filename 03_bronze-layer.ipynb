## Camada Bronze – Ingestão dos Dados CEAP

Responsável pela ingestão dos dados brutos da CEAP, conforme disponibilizados pela Câmara dos Deputados, preservando o formato e a granularidade original.


```python
spark.sql("USE CATALOG mvp_ed_ceap")
spark.sql("USE SCHEMA layer_bronze")
```

```python
df_bronze = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .option("sep", ";")
    .csv("dbfs:/Volumes/mvp_ed_ceap/staging/staging_data/Ano-2024.csv")
)

display(df_bronze.limit(10))

```

```python
(
    df_bronze
    .write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("bronze_ceap_despesas")
)
```
